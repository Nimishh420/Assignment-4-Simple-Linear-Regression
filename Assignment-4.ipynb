```python
dataset=pd.read_csv('delivery_time.csv')
dataset
```

```python
dataset.info()
```

```python
data1=dataset.rename(columns ={'Delivery Time':'dt','Sorting Time':'st'}, inplace =False)
```

```python
sns.distplot(data1['st'])
```

```python
sns.distplot(data1['dt'])
```

```python
data1.corr()
```

```python
sns.regplot(x=data1['st'],y=data1['dt'])
```

```python
#Creating SLR Model -Delivery Time
import statsmodels.formula.api as smf
Delivery_Time = smf.ols("dt~st",data=data1).fit()
```

```python
Delivery_Time.params
```

```python
test_data = pd.DataFrame({'st': [7,10,3,4,2,5]})
test_data
```

```python
Delivery_Time.predict(test_data)
```

```python
#Creating SLR Model-Salary Hike
import statsmodels.formula.api as smf
Salary_hike= smf.ols("Salary~YearsExperience",data=dataset1).fit()
```
